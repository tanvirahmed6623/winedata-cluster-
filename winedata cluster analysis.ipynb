## Objective Function

```python
wcss = [ ]
cluster = range(1,13)

for k in cluster:
    km1 = KMeans(n_clusters=k)
    km1.fit(df)
    wcss.append(km1.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,13),wcss, marker='o')
```

```python
from kneed import KneeLocator as kn
kl = kn(cluster, wcss, direction = 'decreasing', curve='convex')
```

```python
kl.plot_knee()
```

```python
kl.knee
```
